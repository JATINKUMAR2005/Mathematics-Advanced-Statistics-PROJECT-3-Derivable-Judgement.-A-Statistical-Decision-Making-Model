# PART A

## 1. Inferential Statistics

**Definition:**  
Inferential statistics is a branch of statistics that focuses on drawing conclusions about a population using data collected from a sample.  
It allows researchers to estimate population parameters, test hypotheses, and make predictions while accounting for uncertainty.  
Common techniques include confidence intervals, hypothesis testing, and regression analysis.

```python
import numpy as np

sample = np.array([45, 50, 55, 60, 65])
np.mean(sample)
```

## 2. Hypothesis Testing

**Definition:**  
Hypothesis testing is a systematic statistical procedure used to evaluate assumptions or claims about a population parameter.  
It involves formulating a null hypothesis and an alternative hypothesis, calculating a test statistic, and making decisions using p-values or critical values.  
This method helps determine whether observed results are statistically significant or due to random chance.

```python
from scipy import stats

sample = [52, 55, 57, 58, 60]
stats.ttest_1samp(sample, popmean=50)
```

## 3. Confidence Interval and Critical Value

**Definition:**  
A confidence interval is a range of values that is likely to contain the true population parameter with a specified level of confidence.  
The critical value defines the boundary for accepting or rejecting the null hypothesis based on the chosen confidence level.  
Together, they provide a measure of estimation accuracy and reliability.

```python
import numpy as np
from scipy.stats import norm

data = np.array([10, 12, 14, 15, 16])
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z_critical = norm.ppf(0.975)
(mean - z_critical * (std / np.sqrt(n)),
 mean + z_critical * (std / np.sqrt(n)))
```

## 4. p-value

**Definition:**  
The p-value represents the probability of obtaining results as extreme as the observed data, assuming the null hypothesis is true.  
It is used to measure the strength of evidence against the null hypothesis.  
A smaller p-value indicates stronger evidence that the observed result is not due to random chance.

```python
p_value = stats.ttest_1samp(sample, popmean=50).pvalue
p_value
```

## 5. Type I and Type II Errors

**Definition:**  
A Type I error occurs when a true null hypothesis is incorrectly rejected, also known as a false positive.  
A Type II error occurs when a false null hypothesis is not rejected, known as a false negative.  
These errors reflect the risks involved in statistical decision-making.


```python
type_I_error = "Reject H0 when H0 is true"
type_II_error = "Fail to reject H0 when H0 is false"

type_I_error, type_II_error
```

## 6. Statistical Tests (z-test, t-test, Chi-square, ANOVA)

**Definition:**  
Statistical tests are used to analyze data and make inferences about populations.  
The z-test and t-test compare sample means, the chi-square test examines relationships between categorical variables, and ANOVA compares means across multiple groups.  
These tests help determine statistical significance in different scenarios.


```python
from scipy.stats import f_oneway

group1 = [23, 25, 27]
group2 = [30, 32, 34]
group3 = [28, 29, 30]

f_oneway(group1, group2, group3)
```

## 7. Covariance

**Definition:**  
Covariance measures how two variables change together and indicates the direction of their linear relationship.  
A positive covariance shows that variables increase together, while a negative covariance indicates one increases as the other decreases.  
However, covariance does not indicate the strength of the relationship.

```python
import numpy as np

x = [2, 4, 6, 8]
y = [1, 3, 5, 7]

np.cov(x, y)
```

## 8. Correlation

**Definition:**  
Correlation quantifies both the strength and direction of the linear relationship between two variables.  
Its value ranges from -1 to +1, where values closer to ±1 indicate a strong relationship.  
Correlation provides a standardized and easily interpretable measure compared to covariance.

```python
np.corrcoef(x, y)
```

# PART B

## 1. Formulate Hypotheses

**Hypothesis 1 (Smoking vs Diabetes):**  
H₀: Smoking has no effect on diabetes prevalence.  
H₁: Smoking affects diabetes prevalence.

**Hypothesis 2 (BMI vs Diabetes):**  
H₀: Mean BMI of diabetic and non-diabetic individuals is the same.  
H₁: Mean BMI of diabetic and non-diabetic individuals is different.

**Null Hypothesis (H₀):**
$$H_0 : \text{No effect / No difference / No association}$$

**Alternative Hypothesis (H₁):**
$$H_1 : \text{There is an effect / difference / association}$$

```python
import pandas as pd
import numpy as np
from scipy import stats
df = pd.read_csv("health_dataset.csv")  # make sure the CSV is in the same folder
df.head()
```

## 2. Confidence Interval (Mean)

**When population standard deviation is unknown (t-distribution):**
$$CI = \bar{x} \pm t_{\alpha/2,\,n-1}\left(\frac{s}{\sqrt{n}}\right)$$

**When population standard deviation is known (z-distribution):**
$$CI = \bar{x} \pm Z_{\alpha/2}\left(\frac{\sigma}{\sqrt{n}}\right)$$

Where:  
$\bar{x}$ = sample mean  
$s$ = sample standard deviation  
$\sigma$ = population standard deviation  
$n$ = sample size  

```python
ages = df["age"]

mean_age = np.mean(ages)
std_age = np.std(ages, ddof=1)
n = len(ages)

z_critical = stats.norm.ppf(0.975)

ci_age = (
    mean_age - z_critical * (std_age / np.sqrt(n)),
    mean_age + z_critical * (std_age / np.sqrt(n))
)

ci_age
```

## 3. Critical Value and p-value (BMI vs Diabetes)

**Critical Value Rule:**
$$\text{Reject } H_0 \text{ if } |\text{Test Statistic}| > \text{Critical Value}$$

**p-Value Rule:**
$$\text{Reject } H_0 \text{ if } p < \alpha$$

```python
diabetic_bmi = df[df["diabetes"] == True]["bmi"]
non_diabetic_bmi = df[df["diabetes"] == False]["bmi"]

t_stat, p_value = stats.ttest_ind(diabetic_bmi, non_diabetic_bmi)
t_stat, p_value
```

## 4. Perform t-test (Mean BMI Comparison)

$$t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\dfrac{s_1^2}{n_1} + \dfrac{s_2^2}{n_2}}}$$

Used when population variance is unknown and samples are independent.

```python
alpha = 0.05
decision_ttest = "Reject H0" if p_value < alpha else "Fail to Reject H0"
decision_ttest
```

## 5. Chi-square Test (Smoking Status vs Diabetes)

$$\chi^2 = \sum \frac{(O - E)^2}{E}$$

Where:  
$O$ = observed frequency  
$E$ = expected frequency 

```python
contingency_table = pd.crosstab(df["smoking_status"], df["diabetes"])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

chi2_stat, p_value
```

## 6. ANOVA Test (Age Groups vs Diabetes Rate)

$$F = \frac{\text{Mean Square Between Groups (MSB)}}{\text{Mean Square Within Groups (MSW)}}$$

Where:
$$MSB = \frac{SSB}{k - 1}, \quad MSW = \frac{SSW}{n - k}$$

```python
groups = [
    df[df["age_group"] == group]["glucose_level"]
    for group in df["age_group"].unique()
]

f_stat, p_value = stats.f_oneway(*groups)
f_stat, p_value
```

## 7. Covariance and Correlation (Age vs BMI)

$$\text{Cov}(X,Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n - 1}$$

## Correlation (Pearson)

$$r = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}$$

Where:
$$-1 \le r \le 1$$

```python
covariance = np.cov(df["age"], df["bmi"])[0, 1]
correlation = np.corrcoef(df["age"], df["bmi"])[0, 1]

covariance, correlation
```

## 8. Final Results & Interpretation

• Smoking vs Diabetes (Chi-square):  
Reject H₀ if p-value < 0.05 → Smoking affects diabetes prevalence.

• BMI vs Diabetes (t-test):  
Reject H₀ if p-value < 0.05 → Mean BMI differs significantly.

• Age Groups vs Glucose (ANOVA):  
Reject H₀ if p-value < 0.05 → At least one age group differs.

• Age vs BMI:  
Positive correlation indicates BMI tends to increase with age.
